```python
#check if the lines in cleaned file missed any lines
def cleaned_vs_og(cleaned, og, chro_num):
    search_chro = "chr"+str(chro_num)+"$"
    clean_cnt, og_cnt = 0, 0
    with open(cleaned) as f1:
        lines = f1.readlines()
        for i, line in enumerate(lines):
            line_list = line.split()
            is_desired_data = re.match(search_chro, line_list[0]) and bool(re.match("gene", line_list[2]))
            if bool(is_desired_data):
                clean_cnt += 1
    with open(og) as f2:
        lines = f2.readlines()
        for i, line in enumerate(lines):
            line_list = line.split()
            is_desired_data = re.match(search_chro, line_list[0]) and bool(re.match("gene", line_list[2]))
            if bool(is_desired_data):
                og_cnt += 1
    print(clean_cnt, og_cnt)
    return clean_cnt == og_cnt
    
```
